```python
def show_question(path,keys=None, idx=None):
    dataset_questions79 = load_from_disk(f"outputs_test08-checkpoints/{path}")
    if idx is None:
        for ds in dataset_questions79:
            if keys  is None:
                print(ds)
                continue
            for key in keys:
                print(f"{key}: {ds[key]}")
                if key == "oracle_context":
                    print(len(ds[key])), print(len(ds["context"]["sentences"][0]))
                if key == "chunk":
                    print(len(ds[key]))
    else:
        if keys  is None:
            print(dataset_questions79[idx])
            return None
        for key in keys:
                print(f"{key}: {dataset_questions79[idx][key]}")
                if key == "oracle_context":
                    print(len(dataset_questions79[idx][key])), print(len(dataset_questions79[idx]["context"]["sentences"][0]))
```

```python
# show_question("questions/checkpoint-8",idx=1)
# show_question("answers/checkpoint-8", ["question", "oracle_context", "cot_answer"])
show_question("answers/checkpoint-8", ["instruction", "cot_answer"], idx=2)
```

```python
def show_dataset(idx, path):
    dataset_questions = load_from_disk(f"outputs_test04-checkpoints/{path}")
    print(f"type: {type(dataset_questions)}")
    dataset_questions = dataset_questions.to_list()
    # 去掉 dataset_questions 中的 context key
    dataset_questions[idx]["context"] = None
    dataset_questions[idx]["instruction"] = None
    pprint(dataset_questions[idx])
```
